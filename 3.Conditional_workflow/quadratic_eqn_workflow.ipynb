```python
from langgraph.graph import StateGraph , START , END
from typing import TypedDict , Literal
```

```python
#state
class QuadState(TypedDict):
    a : int
    b : int
    c : int

    show_eqn : str
    discriminant : int
    final_answer : str
```

```python
def show_eqn(state : QuadState) -> QuadState :
    result = f"{state['a']}x2+{state['b']}x+{state['c']}"

    return {'show_eqn' : result}
```

```python
def cal_discriminant(state : QuadState) -> QuadState:
    discriminant = state['b']**2 - (4*state['a']*state['c'])

    return {'discriminant' : discriminant}
```

```python
def real_roots(state : QuadState) -> QuadState:
    discriminant = state['discriminant']

    root1 = ((-state['b']) + (discriminant**0.5))/(2*state['a'])
    root2 = ((-state['b']) - (discriminant**0.5))/(2*state['a'])

    output = f"The real roots are {root1} and {root2}"

    return {'final_answer' : output}
```

```python
def repeated_roots(state : QuadState) -> QuadState:
    discriminant = state['discriminant']

    root1 = ((-state['b']) + (discriminant**0.5))/(2*state['a'])

    output = f"The only repeated root is {root1}"

    return {'final_answer' : output}
```

```python
def no_real_roots(state : QuadState) -> QuadState:

    output = f"No real roots as discriminant is negative"

    return {'final_answer' : output}
```

```python
#function to route for conditional nodes
def check_condition(state : QuadState) -> Literal['repeated_roots' , 'no_real_roots' , 'real_roots']:
    disc = state['discriminant']

    if(disc < 0):
        return 'no_real_roots'
    elif(disc > 0):
        return 'real_roots'
    else:
        return 'repeated_roots'
```

```python
#graph
graph = StateGraph(QuadState)

graph.add_node('show_eqn' , show_eqn)
graph.add_node('cal_discriminant' , cal_discriminant)
graph.add_node('no_real_roots' , no_real_roots)
graph.add_node('real_roots' , real_roots)
graph.add_node('repeated_roots' , repeated_roots)

graph.add_edge(START , 'show_eqn')
graph.add_edge('show_eqn' , 'cal_discriminant')
graph.add_conditional_edges('cal_discriminant' , check_condition)
graph.add_edge('no_real_roots' , END)
graph.add_edge('real_roots' , END)
graph.add_edge('repeated_roots' , END)
```

```python
workflow = graph.compile()

initial_state = {
    'a' : 2,
    'b' : 4,
    'c' : 2
}

workflow.invoke(initial_state)
```
